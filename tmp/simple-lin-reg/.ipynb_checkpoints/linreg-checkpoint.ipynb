```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import numpy as np
```

```python
# y = ax + b, straight line fit to data

rng = np.random.RandomState(1) # seed
x = 10 * rng.rand(50) # get 50 nums + scale by 10
y = 2 * x - 5 + rng.randn(50) # linear relationship between x, y (m = 2, b = -5), the add error
plt.scatter(x, y) # plot
```
